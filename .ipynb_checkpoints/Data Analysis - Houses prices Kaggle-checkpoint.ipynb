```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
pd.pandas.set_option('display.max_columns', None)
```

```python
# Load data
data = pd.read_csv('houseprice.csv')

print("The shape of the dataframe is: ", data.shape)
```

```python
data.head()
```

## Missing values

```python
#Make a list of the variables that contain missing values
vars_with_na = [var for var in data.columns if data[var].isnull().sum()>1]

# Print the variables name and percentage of missing values
for var in vars_with_na:
    print(var, np.round(data[var].isnull().mean(), 3), '% missing values')
```

### Relationship between values being missing and Sale Price

```python
def analyse_na_value(df, var):
    df = df.copy()
    
    #Let's make a variable that indicates 1 if the observation was missing or zero otherwise
    df[var] = np.where(df[var].isnull(), 1, 0)
    
    #LEt's calculate the mean SalePrice where the information is missing or present
    df.groupby(var)['SalePrice'].median().plot.bar()
    plt.title(var)
    plt.show()
    
for var in vars_with_na:
    analyse_na_value(data, var)
```

```python
# Let's go ahead and find out what numerical variables we have in the dataset
num_vars = [var for var in data.columns if data[var].dtype != 'O']

print("Number of numerical variables: ", len(num_vars))

#Visualize the numerical variables
data[num_vars].head()
```

### Temporal variables


```python
# List of variables that contain year information
year_vars = [var for var in num_vars if 'Yr' in var or 'Year' in var]
year_vars
```

```python
for var in year_vars:
    print(var, data[var].unique())
    print()
```

```python
data.groupby('YrSold')['SalePrice'].median().plot()
plt.ylabel('Median House Price')
plt.title('Change in House price with the years')
```

This is a very anusual behavior for house prices as they tend to increase during the years

```python
# Let's explore the relationship between the year variables and the house price in a bit of more details
def analyse_year_vars(df, var):
    df = df.copy()
    
    # capture difference between year variable and year the house was sold
    df[var] = df['YrSold'] - df[var]
    
    plt.scatter(df[var], df['SalePrice'])
    plt.ylabel('SalePrice')
    plt.xlabel(var)
    plt.show()
    
for var in year_vars:
    if var != 'YrSold':
        analyse_year_vars(data, var)
```

### Discrete variables 
Let's go agead and find which variables are discrete, i.e., show a finite number of values

```python
# List of discrete variables
discrete_vars = [var for var in num_vars if len(data[var].unique())<20 and var not in year_vars+['Id']]

print('Number of discrete variables: ', len(discrete_vars))
```

```python
# Let's visualize this variables:
data[discrete_vars].head()
```

We can see that htese ariables tend be qualifications of grading scales, or refer to the number of rooms, or units. Let's go ahead and analyse their contribution to the house price

```python
def analyse_discrete(df, var):
    df = df.copy()
    df.groupby(var)['SalePrice'].median().plot.bar()
    plt.title(var)
    plt.ylabel('SalePrice')
    plt.show()
    
for var in discrete_vars:
    analyse_discrete(data, var)
```

### Continuous variables

```python
# List of countinuous variables
cont_vars = [var for var in num_vars if var not in discrete_vars + year_vars + ['Id']]

print('Number of continuous variables: ', len(cont_vars))
```

```python
data[cont_vars].head()
```

```python
#Let's anyalyse the distribution
def analyse_continuous(df, var):
    df = df.copy()
    df[var].hist(bins=20)
    plt.ylabel('Number of houses')
    plt.xlabel(var)
    plt.title(var)
    plt.show()
    
for var in cont_vars:
    analyse_continuous(data, var)
```

```python
# Let's go ahead and analyse the distributions of these variables
def analyse_transformed_continous(df, var):
    df = df.copy()
    
    # Log does not take negative values, so let's be careful and skip those variables
    if 0 in data[var].unique():
        pass
    else:
        # Log transform the variable
        df[var] = np.log(df[var])
        df[var].hist(bins=20)
        plt.ylabel('Number of houses')
        plt.xlabel(var)
        plt.title(var)
        plt.show()
        
for var in cont_vars:
    analyse_transformed_continous(data, var)
```

```python
# Let's now explore the relationship between the house price and transformed variables with more detail

def transform_analyse_continous(df, var):
    df = df.copy()
    
    #Log does not take negative values, so let's be caregul and skip those variables
    if 0 in data[var].unique():
        pass
    else:
        df[var] = np.log(df[var])
        df['SalePrice'] = np.log(df['SalePrice'])
        plt.scatter(df[var], df['SalePrice'])
        plt.ylabel('SalePrice')
        plt.xlabel(var)
        plt.show()
        
for var in cont_vars:
    if var != 'SalePrice':
        transform_analyse_continous(data, var)
```

We can see that 1stFlrSg and GrLivArea have a kind of linear relationship with SalePrice

### Outliers

```python
#Let's make boxplot to visualize outliers in the continuous variables
def find_outliers(df, var):
    df = df.copy()
    
    if 0 in data[var].unique():
        pass
    else:
        df[var] = np.log(df[var])
        df.boxplot(column=var)
        plt.title(var)
        plt.ylabel(var)
        plt.show()
        
for var in cont_vars:
    find_outliers(data, var)
```

We want to ignore the outliers

```python
#Let's visualize the values of categorical variables
data[cat_vars].head()
```

### Rare labels

Let's go ahead and investigate now if there are labels that are present only in a small number of houses

```python
def analyse_rare_labels(df, var, rare_perc):
    df = df.copy()
    tmp = df.groupby(var)['SalePrice'].count() / len(df)
    return tmp[tmp<rare_perc]

for var in cat_vars:
    print(analyse_rare_labels(data, var, 0.01))
```
